```python
addresses = df.current_city.value_counts().index
```

```python
import pandas as pd
from geopy.geocoders import Nominatim

class FarmGeocoder:
    def __init__(self):
        self.geocoder = Nominatim(user_agent="FarmGeocoder")

    def geocode_address(self, address):
        print(address)
        location = self.geocoder.geocode(address)
        if location is not None:
            latitude = location.latitude
            longitude = location.longitude
            return latitude, longitude
        else:
            return None

if __name__ == "__main__":
    # Your addresses list
    addresses = addresses

    # Create a DataFrame
    df = pd.DataFrame({"Address": addresses})

    # Create a new instance of FarmGeocoder
    geocoder = FarmGeocoder()

    # Geocode addresses and create new columns for latitude and longitude
    df[['Latitude', 'Longitude']] = df['Address'].apply(lambda x: pd.Series(geocoder.geocode_address(x) if geocoder.geocode_address(x) else [None, None]))

    # Save the DataFrame to a new Excel file
    df.to_excel("geocoded_addresses.xlsx", index=False)
```
